# Overview
Machine learning is used primarily for solving two classes of problems: classification problems and regression problems. Classification problems involve sorting data into useful sets, and regression involves fitting data to some function. Both of these problem types are common in physics research, and machine learning has been applied to great effect in some cases.
# Classification problems
Machine learning has been used for a variety of classification problems in physics. One notable example is finding states of matter and phase transitions. This has been used very successfully on the Ising model, a simple model for spin-1/2 particles in a magnetic field. At a certain temperature $T_c$, the spins transition from behaving like a paramagnet to behaving like a ferromagnet. Two researchers named Carrasquilla and Melko used a supervised learning algorithm to train a computer to determine whether a given set of spins was in a ferromagnetic or paramagnetic state, and the algorithm was able to determine $T_c$ and the critical exponents that physicists use to study phase transitions. Another researcher named Lei Wang used an unsupervised learning algorithm to do the same.

Machine learning is also used by large experiments such as the LHC use machine learning algorithms to classify the huge amounts of data they bring in, and to determine which signals are worth investigating and might match different particles.
Of immediate relevance to my research is the use of machine learning in neutrino experiments, such as the IceCube observatory at the south pole. Researchers there have used neural networks extensively to classify signals from their detector systems.
It is possible that machine learning systems could also be used to classify different solution sets in my research specifically. I am investigating the initial conditions of proto-neutron stars that lead to the emergence of neutrino driven winds from the stellar surface. It may be that different classes of neutron stars emerge from the data I produce, and machine learning could make classifying those solution sets much easier.
# Regression problems
One of machine learning's greatest strengths is the ability of neural networks to represent any function. This means that very complex data can be accurately fitted regardless of the complexity of the function that describes it. Mathematicians Raissi and Karniadakis have used a machine learning technique to derive several of the fundamental equations describing system behaviors using data directly. They have recovered the nonlinear Schrodinger equation, the Navier-Stokes equations, and several other significant complicated differential equations from data sets using a "hidden physics model" that learns from a data set. 

The equations describing the system I am researching are known fairly well, so I likely will not need to apply regression-type machine learning to my project. I have used algorithms related to machine learning in previous research projects, however. My previous research project involved exploring a number of algorithms commonly used in physics research, including stochastic methods like simulated annealing, and also genetic algorithms. As I understand it, these can be considered forms of machine learning. I also used a newer algorithm known as "Divide and concur," which most nearly resembles a classification type algorithm. It relies on a difference map, and iteratively approaches a resolution of a highly constrained system (e.g. Sudoku puzzle).

# Conclusion
Machine learning is a well-known and well-used tool in physics research. It's possible that I will have occasion to apply it to my current research project to classify different families of solutions to the equations I'm using, but I do not expect to use any regression-type machine learning at the moment.
